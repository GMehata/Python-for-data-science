```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
#Functional Method
```

```python
plt.plot(y)
plt.xlabel('X-axis>')
plt.ylabel('Y-axis')
plt.title('Y=X**2')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
#Object-Oriented Method
#add_axes(rect, projection=None, polar=False, **kwargs)
#rect= [left, bottom, width, height]
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.9])

axes.plot(x,y)
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
plt.show()
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r--')
axes2.plot(y,x,'b')

axes1.set_title('x,y')
axes2.set_title('y,x')
plt.show()
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,'ro-')
axes[0].set_title('x,y')

axes[1].plot(y,x,'bo-')
axes[1].set_title('y,x')

plt.tight_layout()
```

```python
fig1,axes = plt.subplots(2,2)
axes[0,0].plot(x,'g+-')
axes[0,0].set_title('x')

axes[0,1].plot(y,'y+-')
axes[0,1].set_title('y')

axes[1,0].plot(x,y,'ro-')
axes[1,0].set_title('x,y')

axes[1,1].plot(y,x,'bo-')
axes[1,1].set_title('y,x')


plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))

axes[0].plot(x,y,'r-')
axes[0].set_title('x,y')

axes[1].plot(y,x,'b-')
axes[1].set_title('y,x')

plt.tight_layout()
```

```python
fig.savefig('xversusy.png',dpi=300, orientation='landscape')
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0.1,0.1,0.8,0.9])

ax.plot(x,x+2,label='sum2 X')
ax.plot(x,x**2,label='square X')
ax.plot(x,x**3,label='cube X')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.legend(loc=0)
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,0.9,0.8])

ax.plot(x,y,'r', label='XtoY', lw=2, alpha=0.8, ls='-', marker='o', ms=20, mfc='m',mec='g', mew=3)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,0.9,0.8])
ax.plot(x,y,'r', label='XtoY', lw=2, alpha=0.8, ls='--', marker='*', ms=20, mfc='y',mec='g', mew=2)
ax.set_xlim([0,2])
ax.set_ylim([0,3])
```
