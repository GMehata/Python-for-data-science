```python
import seaborn as sns
%matplotlib inline
```

# Matrixplot

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap="YlGnBu" )
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor ='black',linewidths=1)
```

```python
sns.clustermap(fp,cmap='summer',linecolor ='white', linewidths=.25,standard_scale=1,annot=True)
```

# Grids

```python
import matplotlib.pyplot as plt
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
gr=sns.FacetGrid(tips,row='smoker',col='time',margin_titles=True)
gr.map(sns.distplot,'tip',color='c' )
```

```python
g = sns.FacetGrid(tips, col="smoker", row="sex", margin_titles=True)
g = (g.map(plt.scatter, "total_bill", "tip", color="y").set( xlim=(0, 60), ylim=(0, 12),xticks=[10, 30, 50], yticks=[2, 6, 10]).fig.subplots_adjust(wspace=.1, hspace=.2))
```

```python
g = sns.FacetGrid(tips, col="smoker", row="sex", margin_titles=True)
g.map(sns.kdeplot, "total_bill", "tip", color="m",gridsize=8)
```

```python
sns.set_palette('spring', 5, .75)
sns.lmplot(x='total_bill',y ='tip',data=tips, hue= 'sex', markers=['+','o'], scatter_kws={'s':50})
```

```python
sns.set_context('paper',font_scale=2)
sns.lmplot(x='total_bill',y ='tip',data=tips, row='time', col='day', hue= 'sex', markers=['+','o'], palette='husl', scatter_kws={'s':50},)
```

```python
plt.figure(figsize=(10,8))
sns.set_context('paper',font_scale=2)
sns.set_style('whitegrid')
sns.set_palette(['cyan','purple'])
sns.countplot(x='sex',data=tips,hue='smoker')
```
