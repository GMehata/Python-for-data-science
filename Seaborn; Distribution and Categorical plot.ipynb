```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

# Distribution Plot

```python
sns.distplot(tips['total_bill'],kde =False, bins=30, rug= True, rug_kws={"color": "g"}, hist_kws={ "linewidth": 3, "alpha": 1, "color": "c"})
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'scatter',color='m',marker = '^',s=40, edgecolor="y", linewidth=1, marginal_kws=dict(bins=30, rug=True, color='g'))
```

```python
sns.pairplot(tips, hue='sex', palette='husl',kind='scatter', diag_kind='hist', markers='+',plot_kws=dict(s=60, edgecolor="r", linewidth=2))
```

```python
sns.kdeplot(tips['total_bill'],tips['tip'],vertical=True, kernel='tri',bw='silverman', color='m',gridsize=15)
```

# Categorical Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std,)
```

```python
sns.countplot('smoker',data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='total_bill', data= tips, hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill', data= tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data= tips)
```

```python
sns.violinplot(x='day',y='total_bill', data= tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data= tips, hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data= tips, jitter= True)
```

```python
sns.stripplot(x='day',y='total_bill', data= tips, hue='smoker',jitter=True )
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips ,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data= tips, hue='smoker',split=True)
sns.swarmplot(x="day", y="total_bill", data=tips ,hue='sex', palette='husl')
```

```python
sns.catplot(x="sex", y="total_bill",hue="smoker", col="time", data=tips, kind="violin", split=True, height=5, aspect=.8)
```
